```python
import numpy as np
code = """obrhjvaturhftljazqtljbirjzatrjoqhjbvqjiq.hedjlfyftljvapjbqjryrhpyahfrbpjqojbiq.lib--hr-xqtefurhftljryrtbedjurbrhwftftljnhq a fzfbfredatujhrxqtxfzftljirherzodjaejvrzzjaejeirjxq.zudjbqjajxiatlrjeqje.uurtjatueqjfwnqhbatbdjoabfl.rdjatujajhrxqzzrxbfqtjqojirhjzqtlja ertxrdjwaurirhjabjzrtlbijhrb.htjiqwr;jatujeirjrtbrhrujbirjiq.erjvfbijbirjvfeiqojannrahftljxirrho.zjaej.e.azdjatujbirjhreqz.bfqtjqojhrnhreeftlje.xihrozrxbfqtejaejw.ebjwa,rjirhj.tofbjoqhjxqtyrheabfqtgeirjvaejfwwrufabrzpjbqzujbiabjbirjbvqjlrtbzrwrtjohqw"""
txt = "abcdefghijklmnopqrstuvwxyz .,"

def count_letter(code):
    """
    统计各字母出现频率
    :param code: 密文
    :return: 出现次数前三的字符
    """
    letter_dict = {}
    for letter in code:
        try:
            letter_dict[letter] += 1
        except:
            letter_dict[letter] = 1
    result = sorted(letter_dict.items(), key=lambda x: x[1], reverse=True)[:3]
    return result

def get_key(top_letter):
    """
    字幕频率攻击
    :param top_letter: 密文中出现次数最多的3个字符
    :return: [a, b]
    """
    x1 = txt.find('e')+1
    x2 = txt.find('t')+1

    a_list = [0] + list(range(2, 29))
    b_list = list(range(29))
    key = []
    for a in a_list:
        for b in b_list:
            if (x1*a + b) % 29 == txt.find(top_letter[1][0]) and (x2*a + b) % 29 == txt.find(top_letter[2][0]):
                key = [a, b]
    return key

def get_reverse(m, n):
    """
    求乘法逆元
    :param m:
    :param n:
    :return: m^(-1)
    """
    tmp_list = []
    a = m if m>n else n
    b = m + n - a
    while a%b != 0:
        tmp = (-a + a % b) / b
        tmp_list.insert(0 ,tmp)
        mod = a % b
        if mod > b:
            a = mod
        else:
            a = b
            b = mod
    tmp_list.insert(0, (-a + a % b) / b)
    metrix_list = [np.array([[0, 1], [1, i]]) for i in tmp_list]
    result = np.eye(2)
    for metrix in metrix_list:
        tmp = result@metrix
        result = tmp
    return result[0][1]

def get_origin(code, key):
    a_reverse = int(get_reverse(key[0], 29))
    b = key[1]
    result = ""
    for letter in code:
        if letter in txt:
            origin = (a_reverse*txt.find(letter) - a_reverse*b) % 29 - 1
            origin = txt[origin]
        else:
            origin = letter
        result += origin
    return result

top_letter = count_letter(code)
key = get_key(top_letter)
print("key = ", key)
origin = get_origin(code, key)
print(origin)
```

```python
def virginiaCrack(cipherText): # 解密函数
    length = getKeyLen(cipherText) #得到密钥长度
    key = getKey(cipherText,length) #找到密钥
    keyStr = ''
    for k in key:
        keyStr+=k
    print('the key:',keyStr)
    plainText = ''
    index = 0
    c = ''
    for ch in cipherText:
        if ch.islower():
            c = chr((ord(ch)-ord(key[index%length]))%26 + ord('a'))
        elif ch.isupper():
            c = chr((ord(ch)-ord('a')-ord(key[index%length]))%26 + ord('a'))
        else:
            c = ch
        plainText += c
        index+=1
    return plainText

def getKeyLen(cipherText): # 获取密钥长度
    keylength = 1
    maxCount = 0
    for step in range(2,13): # 循环密钥长度
        count = 0
        for i in range(step,len(cipherText)-step):
            if cipherText[i] == cipherText[i+step]:
                 count += 1
        if count>maxCount:
            maxCount = count
            keylength = step
    return keylength

def getKey(text,length): # 获取密钥
    key = [] # 定义空白列表用来存密钥
    alphaRate =[0.08167,0.01492,0.02782,0.04253,0.12705,0.02228,0.02015,0.06094,0.06996,0.00153,0.00772,0.04025,0.02406,0.06749,0.07507,0.01929,0.0009,0.05987,0.06327,0.09056,0.02758,0.00978,0.02360,0.0015,0.01974,0.00074]
    matrix =textToList(text,length)
    for i in range(length):
        w = [row[i] for row in matrix] #获取每组密文
        li = countList(w) 
        powLi = [] #算乘积
        for j in range(26):
            Sum = 0.0
            for k in range(26):
                Sum += alphaRate[k]*li[k]
            powLi.append(Sum)
            li = li[1:]+li[:1]#循环移位
        Abs = 100
        ch = ''
        for j in range(len(powLi)):
             if abs(powLi[j] -0.065546)<Abs: # 找出最接近英文字母重合指数的项
                 Abs = abs(powLi[j] -0.065546) # 保存最接近的距离，作为下次比较的基准
                 ch = chr(j+97)
        key.append(ch)
    return key                

def countList(lis): # 统计字母频度
    li = []
#     tmp = lis.lower()
    alphabet = [chr(i) for i in range(97,123)]
    for c in alphabet:
        count = 0
        for ch in lis:
            if ch.lower() == c:
                count+=1
        li.append(count/len(lis))
    return li

def textToList(text,length): # 根据密钥长度将密文分组
    textMatrix = []
    row = []
    index = 0
    for ch in text:
        row.append(ch)
        index += 1
        if index % length ==0:
            textMatrix.append(row)
            row = []
    return textMatrix

cipherText = input()
plainText= virginiaCrack(cipherText)
plainText
```

```python
def egcd(a, b):
	if a == 0:
		return (b, 0, 1)
	else:
		g, y, x = egcd(b % a, a)
		return (g, x - (b // a) * y, y)
def modinv(a, m):
	g, x, y = egcd(a, m)
	if g != 1:
		raise Exception('modular inverse does not exist')
	else:
		return x % m
    
d=modinv(e,(p-1)*(q-1))
m=pow(c,d,n)
m
```

```python
import socket
from datetime import datetime
from multiprocessing.dummy import Pool as ThreadPool

remote_server = input("Enter a remote host to scan:")
remote_server_ip = socket.gethostbyname(remote_server)
ports = []
print('-' * 60)
print('Please wait, scanning remote host ', remote_server_ip)
print('-' * 60)
socket.setdefaulttimeout(0.5)

def scan_tcp_port(port):
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        res = s.connect_ex((remote_server_ip, port))
        if res == 0:
            print('TCP Port {}: OPEN'.format(port))

        s.close()
    except Exception as e:
        print(str(e.message))


def scan_udp_port(port):
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
        res = s.connect_ex((remote_server_ip, port))
        if res == 0:
            print('UDP Port {}: OPEN'.format(port))

        s.close()
    except Exception as e:
        print(str(e.message))



for i in range(1, 1025):
    ports.append(i)

# Check what time the scan started
t1 = datetime.now()

pool = ThreadPool(processes=1024)
results = pool.map(scan_tcp_port, ports)
results1 = pool.map(scan_udp_port,ports)
pool.close()
pool.join()

print('Multiprocess Scanning Completed in  ', datetime.now() - t1)
```
